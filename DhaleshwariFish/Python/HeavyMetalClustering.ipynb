```python
Cr = list(zip(df.index, df.Cr))

linkage_data = shc.linkage(Cr, method='ward', metric='euclidean')
shc.dendrogram(linkage_data, orientation="right")

plt.show()
```

```python
Cr = list(zip(df.index, df.Cr))
Cd = list(zip(df.index, df.Cd))
Co = list(zip(df.index, df.Co))
Fe = list(zip(df.index, df.Fe))

Crlinkage = shc.linkage(Cr, method='ward', metric='euclidean')
Cdlinkage = shc.linkage(Cd, method='ward', metric='euclidean')
Colinkage = shc.linkage(Co, method='ward', metric='euclidean')
Felinkage = shc.linkage(Fe, method='ward', metric='euclidean')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(8, 3))
dn1 = shc.dendrogram(Crlinkage, ax=axes[0], above_threshold_color='y',
                           orientation='right')
dn2 = shc.dendrogram(Cdlinkage, ax=axes[1], above_threshold_color='y',
                           orientation='right')
dn3 = shc.dendrogram(Colinkage, ax=axes[2], above_threshold_color='y',
                           orientation='right')
dn3 = shc.dendrogram(Felinkage, ax=axes[3], above_threshold_color='y',
                           orientation='right')
```
