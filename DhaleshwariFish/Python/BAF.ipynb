```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

head = pd.read_csv("data/headBAF.csv")
body = pd.read_csv("data/bodyBAF.csv")

logbody = np.log10(body["Values"])
loghead = np.log10(head["Values"])
loghead = loghead * -1

col = "nipy_spectral"
sns.barplot(x = body["Metals"], y= logbody, hue=body["Species"], palette= col)
sns.barplot(x = head["Metals"], y= loghead, hue=head["Species"], palette= col)
plt.axhline(y = 0, color = "white")
plt.legend(loc='upper right', bbox_to_anchor=(1.233, 0.67), fontsize = 15)
plt.tick_params(axis="both", which = "major", labelsize = 20)
plt.xlabel("")
plt.ylabel("Bioaccumulation Factor", fontsize = 25)
plt.text(-0.3, 3.2, '(a) Body', bbox = {'facecolor': 'white', 'alpha': 1, 'pad': 8}, fontsize= 15)
plt.text(-0.3, -3.2, '(b) Head', bbox = {'facecolor': 'white', 'alpha': 1, 'pad': 8}, fontsize= 15)
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

head = pd.read_csv("data/headBAF.csv")
body = pd.read_csv("data/bodyBAF.csv")

logbody = np.log10(body["Values"])
loghead = np.log10(head["Values"])
loghead = loghead * -1

col = "nipy_spectral"
sns.barplot(x = body["Metals"], y= logbody, hue=body["Species"], palette= col)
sns.barplot(x = head["Metals"], y= loghead, hue=head["Species"], palette= col)
plt.axhline(y = 0, color = "white")
plt.legend(loc='upper right', bbox_to_anchor=(1.23, 0.67), fontsize = 15)
plt.tick_params(axis="both", which = "major", labelsize = 20)
plt.xlabel("")
plt.ylabel("BAF", fontsize = 20)
current_values = plt.gca().get_yticks()
ticks = []
for x in current_values:
    string = "10\u00b" + str(x)
    ticks.append(string)
plt.gca().set_yticklabels(ticks)
plt.show()
```
