```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

param = [ "Temperature","pH","EC","TSS","Turbidity","Salinity","TDS","DO","COD","NH4","NO3","Phosphate",
"Sulfates","K","Ca","Mg","Cl","Br","Cd","Cr","Ni","Cu","Zn","Mn","Hg","As","Pb","THB","TC","FC",
"FS","Vibrio C","E Coli"]

fa = pd.read_csv("../Buriganga/Factor_Analysis_All_Param.csv")
fw = pd.DataFrame(data={
    "Parameters": param,
    "F1": fa["Factor 1 W"],
    "F2": fa["Factor 2 W"],
    "F3": fa["Factor 3 W"],
    
})
fs = pd.DataFrame(data={
    "Parameters": param,
    "F1": fa["Factor 1 S"],
    "F2": fa["Factor 2 S"],
    "F3": fa["Factor 3 S"],
    
})


arrayx = np.array([fw["F1"]])
arrayy = np.array([fw["F2"]])
arrayz = np.array([fw["F3"]])

arrayx = arrayx.flatten()
arrayy = arrayy.flatten()
arrayz = arrayz.flatten()

x = preprocessing.normalize([arrayx])
y = preprocessing.normalize([arrayy])
z = preprocessing.normalize([arrayz])
labels = list(fw["Parameters"])

x_reshape = x.reshape(33,)
y_reshape = y.reshape(33,)
z_reshape = z.reshape(33,)

ax3d = plt.figure().gca(projection='3d')

ax3d.scatter(x_reshape, y_reshape, z_reshape, c="red", s= 1)

for x_label, y_label, z_label, label in zip(x_reshape, y_reshape, z_reshape, labels):
    ax3d.text(x_label, y_label, z_label, label, size=12)

ax3d.set_xlabel("Factor 1 (39.75%)", size = 15)
ax3d.set_ylabel("Factor 2 (16.60%)", size = 15)
ax3d.set_zlabel("Factor 3 (16.18%)", size = 15)
plt.show()
```
