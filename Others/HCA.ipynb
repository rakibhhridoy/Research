```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HCAdata.csv",index_col= False)
```

```python
df
```

```python
g = sns.clustermap(df, method="ward", cmap = "viridis", tree_kws=dict(linewidths=1.5, colors=(0.2, 0.2, 0.4)))
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 20)
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 0)
g.ax_cbar.set_position((.11, .31, .03, .255))
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("HCAdata.csv",index_col= False)

g = sns.clustermap(df, method="ward", cmap = "viridis", tree_kws=dict(linewidths=2, colors=(0.2, 0.2, 0.4)), standard_scale= 1)
g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 21)
g.ax_heatmap.set_yticklabels(g.ax_heatmap.get_ymajorticklabels(), fontsize = 0)
g.ax_cbar.set_position((.11, .31, .03, .255))
plt.show()
```
