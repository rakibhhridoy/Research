```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/OraganicCarbonTitration.csv")
```

```python
df
```

```python
df.info()
```

```python
S1 = 1
V1 = 10 
V2 = df["Reading"][0]

S2 = (S1 * V1) / V2
S2
```

```python
w = 0.5
f = S2


percentage_carbon = ((df.Reading[0] - df.Reading) * f * 0.39) / w
```

```python
percentage_carbon
```

```python
df["PCarbon"] = percentage_carbon
df
```

```python
df.to_csv("data/CarbonPercentage.csv")
```
