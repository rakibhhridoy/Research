````python
df = pd.read_csv("../data/SundarbanData.csv", index_col= False)
````

````python
cols = ['Location', 'BulkDensity', 'Sand', 'Silt', 'Clay',
       'Moisture', 'pH', 'EC', 'OrganicCarbon', 'Nitrogen',
       'CEC', 'Textural Class']
````

````python
df.columns = cols
df.head()
````

````python
sdf = df[["Location", "BulkDensity", "OrganicCarbon"]]
sdf.head()
````

```

        SOCD (g C cm-3) = SBD (g cm-3) * (% OC/100)
        SOC (Mg C ha-1) = SBD (g cm-3) * soil depth interval (cm) * SOCC
```

## Soil Organic Carbon Density - SOCD

````python
socd = sdf.BulkDensity * (sdf.OrganicCarbon / 100)
socd
````

````python
soilDepth = 20
soc = sdf.BulkDensity * soilDepth * sdf.OrganicCarbon
soc
````

````python
sdf["SOCD"], sdf["SOC"] = socd, soc
````
