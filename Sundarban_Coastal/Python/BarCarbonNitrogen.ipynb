```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

kjBar = pd.read_csv("../data/kjBar.csv", index_col= False)

sns.catplot(
    data=kjBar, kind="bar",
    x="depth", y="Val", hue="Param",
    errorbar="sd", palette="dark",legend=False, alpha=.6, height=6
)
plt.xlabel("Soil Depth", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.xticks(fontsize = 13)
plt.yticks(fontsize = 13)
plt.legend(title = "Karamjal", fontsize = 13, loc= "upper right")
plt.show()
```
