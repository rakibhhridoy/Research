```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
```

```python
dfA = pd.read_csv("data/LocationD.csv", index_col=False)
dfA = dfA.drop(columns="Unnamed: 0")
dfA.head()
```

```python
temp = dfA.TempD
pH = dfA.pHD
ec = dfA.ECD
do = dfA.DOD
bod = dfA.BODD
tn = dfA.TND
date = dfA.Date
```

```python
val = []

for v in temp:
    val.append(v)
for v in pH:
    val.append(v)
for v in ec:
    val.append(v)
for v in do:
    val.append(v)
for v in bod:
    val.append(v)
for v in tn:
    val.append(v)
```

```python
len(val)
```

```python
param = []

for _ in range(0, 365):
    param.append("Temperature")
for _ in range(0, 365):
    param.append("pH")
for _ in range(0, 365):
    param.append("EC")
for _ in range(0, 365):
    param.append("DO")
for _ in range(0, 365):
    param.append("BOD")
for _ in range(0, 365):
    param.append("TN")
    
len(param)
```

```python
datel = []

for _ in range(0,6):
    for a in date:
        datel.append(a)
        
len(datel)
```

```python
data = pd.DataFrame({
    "Date": datel,
    "Param": param,
    "Values": val,
})
```

```python
data.head()
```

```python
data.to_csv("data/TimeSeriesDataD.csv", index_label=False)
```
