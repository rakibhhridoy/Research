```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## DO

```python
doA = pd.read_csv('DOA/epochs50/metricsT.csv', index_col= False)
doB = pd.read_csv('DOB/epochs50/metricsT.csv', index_col= False)
doC = pd.read_csv('DOC/epochs50/metricsT.csv', index_col= False)
doD = pd.read_csv('DOD/epochs50/metricsT.csv', index_col= False)
```

```python
loc1 = ["S1" for _ in range (0,28)]
loc2 = ["S2" for _ in range (0,28)]
loc3 = ["S3" for _ in range (0,28)]
loc4 = ["S4" for _ in range (0,28)]
```

```python
loc = []
loc.extend(loc1)
loc.extend(loc2)
loc.extend(loc3)
loc.extend(loc4)

model_cat = []
model_cat.extend(doA["Model Category"])
model_cat.extend(doB["Model Category"])
model_cat.extend(doC["Model Category"])
model_cat.extend(doD["Model Category"])

model = []
model.extend(doA["Model"])
model.extend(doB["Model"])
model.extend(doC["Model"])
model.extend(doD["Model"])


mse = []
mse.extend(doA.MSE)
mse.extend(doB.MSE)
mse.extend(doC.MSE)
mse.extend(doD.MSE)

rmse = []
rmse.extend(doA.RMSE)
rmse.extend(doB.RMSE)
rmse.extend(doC.RMSE)
rmse.extend(doD.RMSE)

mae = []
mae.extend(doA.MAE)
mae.extend(doB.MAE)
mae.extend(doC.MAE)
mae.extend(doD.MAE)

r2 = []
r2.extend(doA.R2)
r2.extend(doB.R2)
r2.extend(doC.R2)
r2.extend(doD.R2)


data = pd.DataFrame({
    "Location" : loc,
    "Model Category": model_cat,
    "Model": model,
    "MSE": mse,
    "RMSE": rmse,
    "MAE": mae,
    "R2": r2
})
```

```python
data
```

```python
mse = data.MSE
mae = data.MAE
r2 = data.R2
rmse = data.RMSE
```

```python
l_mse = ["MSE" for _ in range(0,112)]

msedf = pd.DataFrame({
    "Location": data.Location,
    "Metrics": l_mse,
    "Model Category": data["Model Category"],
    "Model": data.Model,
    "y": mse
})


l_mae = ["MAE" for _ in range(0,112)]

maedf = pd.DataFrame({
    "Location": data.Location,
    "Metrics": l_mae,
    "Model Category": data["Model Category"],
    "Model": data.Model,
    "y": mae
})


l_rmse = ["RMSE" for _ in range(0,112)]

rmsedf = pd.DataFrame({
    "Location": data.Location,
    "Metrics": l_rmse,
    "Model Category": data["Model Category"],
    "Model": data.Model,
    "y": rmse
})


l_r2 = ["R2" for _ in range(0,112)]

r2df = pd.DataFrame({
    "Location": data.Location,
    "Metrics": l_r2,
    "Model Category": data["Model Category"],
    "Model": data.Model,
    "y": r2
})
```

```python
msedf
```

```python
metrics = pd.concat([msedf, rmsedf, maedf, r2df], axis=0)
```

```python
metrics
```

```python
metrics.reset_index(inplace=True)
metrics.drop(columns= "index")
```

```python
metrics.to_csv("DOmetrics.csv", index=False)
```
