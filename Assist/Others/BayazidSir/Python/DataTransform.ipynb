```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

directional = pd.read_csv("../plantDataTBidir.csv", index_col=False)
y = np.arange(9)

fig, axes = plt.subplots(ncols=2, sharey=True)
axes[0].barh(y, directional["Dry Weight (t/ha)"], align='center', color='orange', zorder=10)
axes[0].set(title='Dry Weight (t/ha)')
axes[1].barh(y, directional["Fresh Weight (t/ha)"], align='center', color='blue', zorder=10)
axes[1].set(title='Fresh Weight (t/ha)')

axes[0].invert_xaxis()
axes[0].set(yticks=y, yticklabels=directional.Treatment)
axes[0].yaxis.tick_right()

for ax in axes.flat:
    ax.margins(0.03)
    ax.grid(True)

fig.tight_layout()
fig.subplots_adjust(wspace=0.09)
plt.show()
```
